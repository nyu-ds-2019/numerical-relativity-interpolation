```python
import h5py
import numpy as np
```

```python
hf = h5py.File('../data/Proca_fiducial_scaled_cropped.hdf5', 'r')
```

```python
x = hf.get('Train').get('input')
y = hf.get('Train').get('target')
```

```python
inputs = np.array(x)
outputs = np.array(y)
```

```python
inputs.shape, outputs.shape
```

```python
inputs.mean(), inputs.max(), inputs.min()
```

```python
inputs[0, 0].shape
```

```python
import pandas as pd
```

```python
location_value = []
```

```python
import plotly.express as px
```

```python
threshold = 1
```

```python
for index in range(inputs.shape[0]):
    frame = 0
    for i in range(inputs[index, frame].shape[0]):
        for j in range(inputs[index, frame].shape[1]):
            for k in range(inputs[index, frame].shape[2]):
                location_value.append([i, j, k, inputs[index, frame, i, j, k]])

    df = pd.DataFrame(data = location_value, columns=['x', 'y', 'z', 'value'])

    fig = px.scatter_3d(df[df['value'] > threshold], x='x', y='y', z='z', color='value', range_color = [0, 13])

    fig.write_image(f'/Users/paramshah/Downloads/black_hole_simulation_video/images/{index}.png', width=800, height=800)
    
```
